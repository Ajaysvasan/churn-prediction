```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
le  = LabelEncoder()

features = list(data.columns)
for feature in features:
    data[feature] = le.fit_transform(data[feature])
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X = np.array(data.drop('Attrition',axis=1))
Y = np.array(data["Attrition"])
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)

xTrain,xTest,yTrain,yTest = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
type(xTrain)
```

```python
model = Sequential([
    Dense(64,activation='relu',input_shape = (xTrain.shape[1],)),
    Dense(32,activation='relu'),
    Dense(16,activation='relu'),
    Dense(8,activation='relu'),
    # Dense(4,activation='relu'),
    Dense(1,activation='sigmoid')
])

model.compile(optimizer = 'adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(xTrain,yTrain,epochs=100)
```

```python
loss,accuracy = model.evaluate(xTest,yTest)
print(accuracy)
```

```python
class logisticRegression:
    def __init__(self,layer_dims):
        self.layer_dims = layer_dims
        self.parameters = {}
        # initializing bias and weigths using He initialization
        self.L = len(self.layer_dims) - 1
        for i in range(1,self.L + 1):
            self.parameters[f'W{i}'] = np.random.randn(layer_dims[i],layer_dims[i-1])*np.sqrt(2/self.layer_dims[i-1])
            self.parameters[f'B{i}'] = np.zeros((self.layer_dims[i],1))
    
    def ReLU(self,Z):
        return np.maximum(0,Z)
    
    def Sigmoid(self,Z):
        return 1/(1+np.exp(np.clip(Z,-500,500)))
    
    def derevative_ReLU(self,Z):
        return np.where(Z>0,1,0)
    
    def derevative_Sigmoid(self,Z):
        s = self.Sigmoid(Z)
        return s*(1-s)
    
    def forward_propagation(self,X):
        cache = {}
        A = X.T
        cache['A0'] = A
        for i in range(1,self.L+1):
            Z = np.dot(self.parameters[f'W{i}'],A) + self.parameters[f'B{i}']
            cache[f'Z{i}'] = Z
            A = self.Sigmoid(Z) if i==self.L else self.ReLU(Z)
            cache[f'A{i}']=A
        return A,cache
    
    def computeCost(self,AL,Y):
        m = Y.shape[0]
        Y = Y.reshape(1, m)
        
        # Added epsilon for numerical stability
        epsilon = 1e-15
        AL = np.clip(AL, epsilon, 1 - epsilon)
        
        cost = -(1/m) * np.sum(Y * np.log(AL) + (1 - Y) * np.log(1 - AL))
        
        return np.squeeze(cost)

    def back_propagation(self,cache,Y):
        gradients = {}
        m = Y.shape[0]
        Y = Y.reshape(1,m)
        dAL = -(np.divide(Y,cache[f'A{self.L}'])-np.divide((1-Y),1-cache[f'A{self.L}']))

        for i in reversed(range(1,self.L+1)):
            Z = cache[f'Z{i}']
            A_prev = cache[f'A{i-1}']
            dZ = dAL*self.derevative_Sigmoid(Z) if self.L == i else dAL*self.derevative_ReLU(Z)
            gradients[f'dW{i}'] = (1/m)*np.dot(dZ,A_prev.T)
            gradients[f'dB{i}'] = (1/m)*np.sum(dZ,axis=1,keepdims=True)
            dAL = np.dot(self.parameters[f'W{i}'].T,dZ) if i>1 else None
        return gradients
    
    def updateParameters(self,gradient,lr):
        for i in range(1,self.L+1):
            self.parameters[f'W{i}'] -=  lr*gradient[f'dW{i}']
            self.parameters[f'B{i}'] -=  lr*gradient[f'dB{i}']

    def train(self,X,Y,lr=0.01,epochs=1000,print_cost=True):
        costs = []
        for epoch in range(epochs):
            AL,cache = self.forward_propagation(X)
            cost = self.computeCost(AL,Y)
            gradients = self.back_propagation(cache,Y)
            self.updateParameters(gradient=gradients,lr=lr)
            if print_cost and epoch%100==0:
                print(f"The cost of the model at {epoch}: {cost}")
            costs.append(cost)
        return cost
    def predict(self,X):
        AL,_ = self.forward_propagation(X)
        prediction = (AL>0.5).astype(int)
        return prediction
    
    def evaluate(self,X,Y):
        AL,_ = self.forward_propagation(X)
        Y = Y.reshape(1,Y.shape[0])
        accuracy = np.mean(AL==Y)*100
        return accuracy
    
model = logisticRegression([X.shape[1],64,32,16,8,1])
model.train(xTrain,yTrain)
```

```python
accuracy = model.evaluate(xTest,yTest)
```
